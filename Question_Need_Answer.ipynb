```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Admin\Desktop\BTL_BigData\data\dataset_1761666182405\ecommerce_dataset_updated.csv")

print("===== CÂU 1: Dataset có bao nhiêu dòng và cột? =====")
print(f"Số dòng: {df.shape[0]}")
print(f"Số cột: {df.shape[1]}")

print("\n===== CÂU 2: Có bao nhiêu giá trị bị thiếu hoặc không hợp lệ? =====")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "Không có giá trị thiếu!")

print("\n===== CÂU 3: Các cột quan trọng trong bộ dữ liệu =====")
print(list(df.columns))

print("\n===== CÂU 4: Dữ liệu trải dài trong khoảng thời gian nào? =====")
# Nếu có cột ngày, thử tìm cột có chứa từ khóa date
date_cols = [col for col in df.columns if 'date' in col.lower()]
if date_cols:
    df[date_cols[0]] = pd.to_datetime(df[date_cols[0]], errors='coerce')
    print(f"Khoảng thời gian: từ {df[date_cols[0]].min()} đến {df[date_cols[0]].max()}")
else:
    print("⚠️ Không tìm thấy cột thời gian trong dataset.")

print("\n===== CÂU 5: Có cột nào cần tách, gộp hoặc xử lý đặc biệt không? =====")
print("→ Kiểm tra kiểu dữ liệu:")
print(df.dtypes)
print("\nGợi ý: Kiểm tra các cột có dạng phần trăm (%) hoặc ký hiệu tiền tệ để xử lý.")

print("\nCâu 6: Tự suy nghĩ ra insight và bài toán từ dữ liệu đã phân tích và từ cái insight đó đưa ra bài toán phù hợp với mô hình học máy và ứng dụng vào thực tế?")
```
